```python
import pandas as pd
```

```python
df = pd.read_csv('credit_train_edited_version.csv')
id = '14dd8831-6af5-400b-83ec-68e61888a048'
```

```python
def search_cid(id):
    for i in range(len(df)):
        tmp = df.loc[i].values
#         print(tmp[0])
        if tmp[0] == id:
            return tmp
print(search_cid(id))
```
